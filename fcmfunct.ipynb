```python
import random
import numpy as np
from numpy import linalg as LA

import FCMeans as fcm
import matplotlib.pyplot as plt
from sklearn.datasets.samples_generator import make_blobs
```

```python
# Dados para os problema
mu = 2 #fuzificador expoente peso
c = 4 #qntde de grupos
eps = 1e-10 #criterio de parada
itmax = 500
```

```python
n = 500
X, y= make_blobs(n_samples=n, centers=c, random_state=1, cluster_std=1)
plt.scatter(X[:, 0], X[:, 1], s=30, cmap='autumn')
plt.show()

X1, y1 = make_blobs(n_samples=n, centers=c, random_state=1, cluster_std=15)
plt.scatter(X1[:, 0], X1[:, 1], s=30, cmap='autumn')
plt.show()
```

```python
C1 = fcm.C_init(X,c) # centroides iniciais
C,U,it = fcm.FCM(X,mu,c,eps,itmax,C1)
print(it, 'iterações')
plt.plot(X[:, 0], X[:, 1],'bo')
plt.plot(C[:, 0], C[:, 1],'ro') 
plt.plot(C1[:, 0], C1[:, 1],'yo')
plt.show()
```

```python
C1 = fcm.C_init(X1,c) # centroides iniciais
C,U,it = fcm.FCM(X1,mu,c,eps,itmax,C1)
print(it, 'iterações')
plt.plot(X1[:, 0], X1[:, 1],'bo')
plt.plot(C[:, 0], C[:, 1],'ro') 
plt.plot(C1[:, 0], C1[:, 1],'yo')
plt.show()
```
